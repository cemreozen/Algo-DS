### Aufgabe: Binary Search


#### Problembeschreibung
    
Auf einem sortieren Feld (*Array*) (mit definierter Ordnung) soll ein Element gefunden werden bzw. das Fehlen des Elementes festgestellt werden.


#### Idee des Algorithmus

Rekursiv: 
- Sehe in der Mitte des Feldes nach und vergleiche das Suchelement mit dem Element in der Mitte des Feldes. 
Wenn die beiden Element gleich sind, wurde das Element gefunden (Abbruch der Suche!).
Falls das Suchelement kleiner/größer ist als das Element in der Mitte, wäre das Suchelement (falls vorhanden) in der ersten bzw. zweiten Hälfte. 
- Teile das Feld in zwei Hälften und betrachte wieder das mittlere Element der entsprechenden relevanten Hälfte. Falls das Feld nicht mehr geteilt werden kann, war das Element nicht im ursprünglichen Feld vorhanden (Abbruch!).

Man kann sich die Unterteilungen als Baum vorstellen, wobei der Baum nicht explizit konstruiert wird. Hier ist ersten Ebenen des Baums:


#### Aufgabe (ohne Recherche!)

- Schreibe den Algorithmus als Pseudocode. Im Fall, dass das Element im Feld gefunden wird, soll `true` ausgegeben werden sonst `false`. 
- Implementieren Sie Binary-Search in go unter Verwendung von Rekursionen.

```go
// Erzeuge Array zum Testen 

//rand.Seed(time.Now().UnixNano())
rand.Seed(42) 

func randbool() bool {
    return rand.Float32() < 0.5
}
var s []int // slice
m := 100
for i:=0; i < m; i++ {
    if randbool() {
        s = append(s, i)   
    }
}
s // test your binary-search on this array
```

```go
func BinarySearch(data []int, value int) bool{
        var array_length int = len(data) - 1
        var starting_index int = 0
        var middle int = (starting_index + array_length) / 2
        for starting_index <= array_length {
                 if data[middle] == value {
                         return true
                 }       
                 if data[middle] > value {
                         array_length = middle -1
                 }       
                 if data[middle] < value {
                         starting_index = middle + 1
                 }       
         }       
         return false
 }
```
